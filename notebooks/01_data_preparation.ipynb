# Data Preparation
Prepare raw MRI images for training and evaluation.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
raw_path = "../dataset/raw/Training"
processed_path = "../dataset/processed"

print("Raw Dataset:", raw_path)
print("Processed Dataset:", processed_path)
```

```python
splits = ["train", "val", "test"]

classes = os.listdir(raw_path)

for split in splits:
    for cls in classes:
        os.makedirs(os.path.join(processed_path, split, cls), exist_ok=True)

print("Folders Created Successfully")
```

```python
for cls in classes:

    class_path = os.path.join(raw_path, cls)

    images = os.listdir(class_path)

    train_imgs, temp_imgs = train_test_split(
        images,
        test_size=0.30,
        random_state=42
    )

    val_imgs, test_imgs = train_test_split(
        temp_imgs,
        test_size=0.50,
        random_state=42
    )

    print(f"{cls}")
    print(f"Train: {len(train_imgs)}")
    print(f"Val: {len(val_imgs)}")
    print(f"Test: {len(test_imgs)}")
    print("-"*30)
```

```python
for cls in classes:

    class_path = os.path.join(raw_path, cls)

    images = os.listdir(class_path)

    train_imgs, temp_imgs = train_test_split(
        images,
        test_size=0.30,
        random_state=42
    )

    val_imgs, test_imgs = train_test_split(
        temp_imgs,
        test_size=0.50,
        random_state=42
    )

    for img in train_imgs:
        shutil.copy(
            os.path.join(class_path, img),
            os.path.join(processed_path, "train", cls, img)
        )

    for img in val_imgs:
        shutil.copy(
            os.path.join(class_path, img),
            os.path.join(processed_path, "val", cls, img)
        )

    for img in test_imgs:
        shutil.copy(
            os.path.join(class_path, img),
            os.path.join(processed_path, "test", cls, img)
        )

print("Dataset Split Completed")
```

```python
for split in ["train", "val", "test"]:

    print(f"\n{split.upper()} DATA")

    split_path = os.path.join(processed_path, split)

    for cls in os.listdir(split_path):

        count = len(os.listdir(os.path.join(split_path, cls)))

        print(f"{cls}: {count}")
```
